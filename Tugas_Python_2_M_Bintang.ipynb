```python
import pandas as pd

df = pd.read_csv('data0-SZL.csv')
calculations = {
    "mean": df.mean(),
    "std": df.std(),
    "skew": df.skew(),
    "median": df.median(),
    "energy": df.apply(lambda x: (x**2).sum()**0.5),
    "min": df.min(),
    "max": df.max()
}

results_df = pd.DataFrame(calculations)

results_df
```

```python
# @title median

from matplotlib import pyplot as plt
results_df['median'].plot(kind='line', figsize=(8, 4), title='median')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# @title mean

from matplotlib import pyplot as plt
results_df['mean'].plot(kind='hist', bins=20, title='mean')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract mean and standard deviation for the scatter plot
x = results_df['mean']
y = results_df['std']

# Prepare color and marker lists
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']
markers = ['*', 'o', 'v', '^', '<', '>', 's', 'p', 'h', 'D']

# Plotting
plt.figure(figsize=(10, 6))

# Split data into groups of 10 and plot each with different color and marker
for i in range(0, len(x), 10):
    plt.scatter(x[i:i+10], y[i:i+10], color=colors[(i//10) % len(colors)], marker=markers[(i//10) % len(markers)], label=f'Kolom {i+1}-{i+10}')

plt.title('Scatter RATA-RATA vs. Scatter Deviasi')
plt.xlabel('RATA-RATA')
plt.ylabel('Scatter Deviasi')
plt.legend(title='Column Groups', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

```python

x_median = results_df['median']
y_skewness = results_df['skew']


plt.figure(figsize=(10, 6))
for i in range(0, len(x_median), 10):
    plt.scatter(x_median[i:i+10], y_skewness[i:i+10], color=colors[(i//10) % len(colors)], marker=markers[(i//10) % len(markers)], label=f'Kolom {i+1}-{i+10}')

plt.title('Diagram Titik Median vs. Skewness berdasarkan Grup Kolom')
plt.xlabel('Median')
plt.ylabel('Skewness')
plt.legend(title='Grup Kolom', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```
